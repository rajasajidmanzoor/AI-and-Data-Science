##Easy to use (Simplicity)

```python
l= [{1,2,3},
   {4,5,6,},
   {7,8,9}]

l
```

```python
for rid,row in enumerate(l):
    print(rid,row)
    for cid,_ in enumerate(row):
        print(cid, end="***")
    print()
```

```python
for rid,row in enumerate(l):
    for cid,_ in enumerate(row):
        l[rid][cid] += 5
l
```
